Using weather API to get historical data

Using wunderground created issues since its table is dynamically created using JS, and thus cant be accessed through DOM. And I didnt want to use selenium

API Endpoint is: https://api.weather.com/v1/location/KPHL:9:US/observations/historical.json?apiKey=e1f10a1e78da46f5b10a1e78da96f525&units=e&startDate=\<date\>

```python
import requests
from datetime import datetime


base_url = 'https://api.weather.com/v1/location/KPHL:9:US/observations/historical.json?apiKey=e1f10a1e78da46f5b10a1e78da96f525&units=e&startDate='
phl_weather = []

# Get the average temp for each date
for date in dates:
  full_url = base_url + date
  print(date)
  try:
      response = requests.get(full_url).json()['observations']
      total = 0
      num_obs = 0
      for obs in response:
        if obs['temp'] != None:
          total += obs['temp']
          num_obs += 1
      avg_temp = total / num_obs
      phl_weather.append([str(datetime.strptime(date, '%Y%m%d').date()), avg_temp])        
  except:
      pass
        
# Convert to a dataframe
cols = ['Date', 'Avg Temp']
phl_weather = pd.DataFrame(phl_weather, columns=cols)
        

```
